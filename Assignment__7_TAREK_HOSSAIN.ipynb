```python
#homework
class Employee:
    employee_count = {}

    def __init__(self, name, joining_date, work_experience, weekly_work_hour = 40):
        self.name = name
        self.joining_date = joining_date
        self.work_experience = work_experience
        self.weekly_work_hour = weekly_work_hour
        if self.weekly_work_hour > 60 or self.weekly_work_hour < 40:
            print(f"{self.name} can not work for {weekly_work_hour} hours.")
            self.weekly_work_hour = 40

    @classmethod
    def showDetails(cls):
        total_employee_count = sum(cls.employee_count.values())
        print(f"Company workforce: \nTotal Employee/s: {total_employee_count}")

        if total_employee_count > 0:
           print(f"Total Programmer/s: {cls.employee_count['Programmers']}")
           print(f"Total HR programmer/s: {cls.employee_count['HR']}")

class Programmer(Employee):
    designation_list = []

    def __init__(self, name, joining_date, work_experience, weekly_work_hour = 40, status = False):
        super().__init__(name, joining_date, work_experience, weekly_work_hour)
        if work_experience > 0 or status == True:
          super().employee_count["Programmers"] = super().employee_count.get("Programmers", 0) + 1

        self.id = self.createProgrammerID()
        self.designation = ""
        if self.work_experience >= 0 and self.work_experience < 3:
            self.designation = "Junior Software Engineer"
        elif self.work_experience >= 3 and self.work_experience < 5:
            self.designation = "Software Engineer"
        elif self.work_experience >= 5 and self.work_experience < 8:
            self.designation = "Senior Software Engineer"
        else:
          self.designation = "Technical Lead"

        Programmer.designation_list.append(self.designation)
        self.salary = self.calculateSalary()

    def showProgrammerDetails(self):
        joinig_date = self.joining_date.strip().split('-')
        print("Programmer Employee:")
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        print(f"Joining Date: {joinig_date[2]+'-'+joinig_date[1]+'-'+joinig_date[0]}")
        print(f"Designation: {self.designation}")
        print(f"Salary: BDT {self.salary}")

    def calculateSalary(self):
        year = 2023 - int(self.joining_date.strip().split('-')[0])
        salary = 0
        if self.designation == "Junior Software Engineer":
          salary = 30000
        if self.designation == "Software Engineer":
          salary = 45000
        if self.designation == "Senior Software Engineer":
          salary = 70000
        if self.designation == "Technical Lead":
          salary = 120000

        for i in range(year):
            increment = salary * 15/100
            salary += increment

        return salary

    def createProgrammerID(self):
        id = "P-"+self.joining_date[2:4]+self.joining_date[5:7]+self.joining_date[8:]+'-'+str(sum(Employee.employee_count.values()))
        return id

    def calculateOvertime(self):
        if self.weekly_work_hour == 40:
            print(f"{self.name} will not get overtime.")
            return

        overtime = self.weekly_work_hour - 40
        over_time_salary = 500 * overtime * 4
        self.salary += over_time_salary
        print(f"{self.name} will get BDT {over_time_salary} overtime.")

class HR(Employee):
    def __init__(self, name, joining_date, work_experience, weekly_work_hour = 40):
        super().__init__(name, joining_date, work_experience, weekly_work_hour)
        super().employee_count["HR"] = Employee.employee_count.get("HR", 0) + 1
        self.id = self.createHREmployeeID()

    def showHREmployeeDetails(self):
        joiningdate = self.joining_date.strip().split('-')
        print("HR Employee:")
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        print(f"Joining Date: {joiningdate[2] + '-' + joiningdate[1] + '-' + joiningdate[0]}")

    def createHREmployeeID(self):
        id = "HR-" + self.joining_date[2:4] + self.joining_date[5:7] + self.joining_date[8:] + '-' + str(sum(Employee.employee_count.values()))
        return id

class InternProgrammer(Programmer):
    intern_count = 0

    def __init__(self, name, joining_date, intern_type = "Unpaid"):
        super().__init__(name, joining_date, 0, 40)
        InternProgrammer.intern_count += 1
        self.intern_type = intern_type
        self.id = f"Temp_{str(InternProgrammer.intern_count)}"
        joiningdate = self.joining_date.strip().split('-')
        self.status = False
        if joiningdate[1] == "01" or joiningdate[1] == "07":
            self.status = True

    def showInternDetails(self):
        print("Intern (Programmer):")
        print(f"Name: {self.name}")
        print(f"ID: {self.id}")
        print(f"Joining Date : {self.joining_date}")
        print(f"Type: {self.intern_type}")
        if self.status==True:
           print("Status: Eligible for promotion")
        else:
           print("Status: Not Eligible for promotion")

    def promoteToProgrammer(self):
        if self.status==True:
            self.joining_date = "2023-08-21"
            newprogrammer = Programmer(self.name, self.joining_date, 0, 40, self.status)
            print(f"{self.name} is promoted!")
            return newprogrammer
        else:
            print(f"{self.name} can not be promoted!")

Employee.showDetails()
print("=========1=========")
richard = Programmer("Richard Hendricks", "2021-06-08", 4, 48)
richard.calculateSalary()
print("=========2=========")
richard.showProgrammerDetails()
print("=========3=========")
richard.calculateOvertime()
print("=========4=========")
richard.showProgrammerDetails()
print("=========5=========")
monica = HR("Monica Hall", "2022-07-06", 2, 40)
print("=========6=========")
monica.showHREmployeeDetails()
print("=========7=========")
Employee.showDetails()
print("=========8=========")
gilfoyle = Programmer("Bertram Gilfoyle", "2020-03-02", 6, 35)
gilfoyle.calculateSalary()
print("=========9=========")
gilfoyle.calculateOvertime()
print("=========10=========")
gilfoyle.showProgrammerDetails()
print("=========11=========")
gavin = Programmer("Gavin Belson", "2016-12-20", 9)
gavin.calculateSalary()
gavin.calculateOvertime()
gavin.showProgrammerDetails()
print("=========12=========")
yang = InternProgrammer("Jian Yang", "2023-01-01")
yang.showInternDetails()
print("=========13=========")
jared = InternProgrammer("Jared Dunn", "2023-06-05", "Paid")
jared.showInternDetails()
print("=========14=========")
jared = jared.promoteToProgrammer()
print("=========15=========")
Employee.showDetails()
print("=========16=========")
yang = yang.promoteToProgrammer()
yang.calculateSalary()
yang.showProgrammerDetails()
print("=========17=========")
Employee.showDetails()
```
