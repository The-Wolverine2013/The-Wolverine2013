```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint

# Nonlinear oscillator parameters
k = 1.0  # Spring constant
m = 1.0  # Mass
gamma = 0.2  # Damping coefficient
alpha = 0.5  # Nonlinear term coefficient

# Differential equation system for the oscillator
def project(y, t):
    #print(y)
    #print("this is y: ",y)
    x, v = y  # Position and velocity
    dx_dt = v
    #print(t)
    #print(v)
    #print(dx_dt)
    dv_dt = -k/m * x - gamma/m * v - alpha/m * x**3  # Nonlinear restoring force
    return [dx_dt, dv_dt]

# Initial conditions: [initial position, initial velocity]
y0 = [1.0, 0.0]

# Time points for integration
t = np.linspace(0, 20, 1000)
#print(t)

# Solve the system of differential equations
sol = odeint(project, y0, t) #y0=initial value of Y; t= the time space for which we want the curve(basically the range of x)
# Extract position and velocity from the solution
x_values = sol[:, 1]
v_values = sol[:, 0]
#print(x_values)
print(len(v_values))
# Visualization
plt.figure(figsize=(15, 7))
plt.plot(t, x_values, label='Position')
plt.plot(t, v_values, label='Velocity')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.title('Nonlinear Oscillator Simulation')
plt.legend()
plt.grid()
plt.show()
```
