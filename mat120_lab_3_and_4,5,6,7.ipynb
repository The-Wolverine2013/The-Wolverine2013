```python
import sympy as sp
from sympy import*
sp.init_printing()
x=sp.symbols('x')
y=x**4 -(2/3)*x**3-2*x**2 +2*x
max=0
min=0
maxima=None
minima=None
maxima_minima_counter=0
l1=[]
expr1=(sp.diff(y,x))
solve1=sp.solve(expr1)
for i in solve1:
  if 0<=i<=3:
    l1.append(i)
expr2=sp.diff(expr1,x)
for j in l1:
  maxima_minima_counter=expr2.subs(x,j)
  if maxima_minima_counter>0:
    min=j
  else:
    max=j
maxima=y.subs(x,max)
minima=y.subs(x,min)
print(f"minima is {minima} and maxima is {maxima}")
```

```python
#impelmentation of the Simpson's 1/3 rule
#importing library functions
import math

#functions to calculate f(x)
def f1(x):
    return math.sin(x)
def f2(x):
    return x**2

#function for approximate integral determination
'''Simpson's 1/3 rule=integral of f(x)dx=(h/3)*[(first+last)+4*(sum of odds)+2*(sum of evens)]'''
def simpsons_(a,b,f,n):
    #calculation of the value of interval
    h = (b-a)/n

    #list for storing value of x and f(x)
    x=[]
    fx=[]

    #calculation of values of x and f(x)
    i=0
    while i<=n:
        x.append(a+i*h)
        fx.append(f(x[i]))
        i+=1

    #calculation of the result
    result=0
    i=0
    while i<=n:
        if i==0 or i==n:
            result+=fx[i]
        elif i%2!=0:
            result+=4*fx[i]
        else:
            result+=2*fx[i]
        i+=1
    result=result*(h/3)
    return result

#showing result for given limit
lower_limit=3
upper_limit=4.8
n=6
r1=simpsons_(lower_limit, upper_limit, f1, n)
r2=simpsons_(lower_limit, upper_limit, f2, n)
print("The value of integration of sin(x) for limit 3 to 4.8 is %.6f" % r1)
print("The value of integration of x^2 for limit 3 to 4.8 is %.6f" % r2)
```
